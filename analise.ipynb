1 - Acessando os dados da tabela.

```python
import pandas as pd

tabela = pd.read_excel('tabela_alunos.xlsx', engine='openpyxl')
display(tabela)
```

2 - Quantidade de alunos presentes na lista.

```python
print(f'Na tabela de alunos tem {len(tabela.index)} alunos.')
```

3 - Controle de entrada e saída de alunos.

```python
registro = tabela[["Nome","Atividade"]].groupby(["Nome"]).sum()
display(registro)
```

4 - Alunos que saíram da aula, mas retornaram.

```python
registro.loc[registro['Atividade'] == 'EntrouSaiuEntrou']
```

5 - Alunos que saíram da aula.

```python
registro.loc[registro['Atividade'] == 'EntrouSaiu']
```
